## KDE Plot

- KDE plot used to represent the density of observations lies
- Same as histogram but difference in representation

### Univariate

```python
plt.figure(figsize=(15,5))
sns.kdeplot(x = df['12graduation'])
plt.show()
```

Between the years 2005 to 2011 there are more graduates of 12th than the remaining years

### Bi Variate

```python
plt.figure(figsize=(15,5))
sns.kdeplot(x = df['12percentage'], y = df['10percentage'])
plt.show()
```

- Percentage obtained by 10 th and 12 th were in between the range 60% to 90%

## ECDF Plot

ECDF Plot is used to represent the entire data from the least to the greatest 

```python
plt.figure(figsize=(15,5))
sns.ecdfplot(x = df['12graduation'])
plt.show()
```

- From 1995 to 2002 graduation rate was constant and least
- and the highest was in 2012 
- X axis represents 12 Graduation and Y for Proportion

## Rug Plot

- Rug Plot represented the distribution of values along the X and Y axes 
- Rug Plot is effective when combined with other type of plot 

```python
plt.figure(figsize=(15,5))
sns.rugplot(x = df['12percentage'])
plt.show()
```

- X axis represents 12 Percentage

```python
plt.figure(figsize=(15,5))
sns.rugplot(x = df['12percentage'])
sns.kdeplot(x = df['12percentage'])
plt.show()
```


- X axis represents 12 Percentage and Y for Density
- High density lies between 60 to approximately 98

```python
plt.figure(figsize=(15,5))
sns.rugplot(x = df['12percentage'], y = df['10percentage'])
plt.show()
```

- X axis represents 12 Percentage and Y for 10 Percentage
- High density lies between 60 to approximately 98 for 12 th Percentage
- High density lies between 50 to approximately 98 for 10 th Percentage


```python
plt.figure(figsize=(15,5))
sns.kdeplot(x = df['12percentage'], y = df['10percentage'])
sns.rugplot(x = df['12percentage'], y = df['10percentage'])
plt.show()
```

# Box Plot
Boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending from the boxes (whiskers) indicating variability outside the upper and lower quartiles,Outliers may be plotted as individual points.

```python
plt.figure(figsize=(15,5))
sns.boxplot(x=df.Salary,)
plt.show()
```

- Median of Salary is at areound 0.4 
- Outliers starts from approximately 0.6

```python
plt.figure(figsize=(15,5))
sns.boxplot(x=df.Salary,y=df.Gender)
plt.show()
```

- Median is almost same for both genders 
- Compared to Females salaries for male is high than female 
- Outliers are in high density for males and most lies between around 0.7 to 1.0

# Bar Plot
## catergorical Data
A bar plot or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.

```python
plt.figure(figsize=(10,5))
df["CollegeState"].value_counts().plot.bar()
plt.show()
```

- States with high number of colleges are as follows 
- UP,Karnataka,TN,Telangana,Maharastra,AP,AB,Punjab,MP,Haryana,Rajastan,Orissa,Delhi 

# Bivariate Scatter plot
Comparing multiple variables simultaneously is also another useful way to understand your data. When you have two continuous variables, a scatter plot is usually used.

```python
df.plot(x='agreeableness',y='extraversion',kind='scatter')
plt.show()
```

# Joint Plot
Join plots can do both univariate as well as bivariate analysis. The main plot will give us a bivariate analysis, whereas on the top and right side we will get univariate plots of both the variables that were considered. It makes our job easy by getting both scatter plots for bivariate and Distribution plot for univariate, both in a single plot.

```python
sns.jointplot(data=df,x='agreeableness',y='extraversion',kind="hex")
plt.show()
```

## Swarmplot

- Swarmplot represented by dots and are arranged in a way that the dots doesn't overlap with other dots
- And is almost similar as Stripplot

```python
sns.swarmplot(x='Degree', y='Salary', hue= "Gender",data=df)
plt.show()
```

- Orange is for Male and Blue is for Female 
- Salary for both the genders are in high density between the level 0.0 to 1.5 for B.Tech/B.E
- 0.0 to 1.0 for MCA
- Around half of MCA is for M.Tech/M.E
- Very least in density is for M.Sc
